# Splitting the dataset into the training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
future=data.drop(['Quality','A_id'],axis=1)
target=data['Quality']
x_train,x_test,y_train,y_test=train_test_split(future,target,test_size=0.2)
```

```python
x_train
```

# Fitting K-NN to the training set

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
model=knn.fit(x_train,y_train)
```

# Predicting the test set results

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

# Machine performance score

```python
model.score(x_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

# Making the confusion matrix

```python
con=confusion_matrix(y_test,y_pred)
```

```python
con
```

# confustion matrix Heatmap

```python
sns.heatmap(con,cmap='Greens',center=0,square=True,annot=True,cbar_kws={'orientation':'vertical','label':'color bar'},xticklabels=['Good','Bad'],yticklabels=['Good','Bad'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Apple Quality')
plt.show()
```
