# Outlier - detection method

```python
a=data.boxplot()
a.set_xticklabels(a.get_xticklabels(),rotation=30,ha='right',fontsize=10)
plt.tight_layout()
```

```python
data['Unit Price']=np.where(data['Unit Price']>=1200,data['Unit Price'].median(),data['Unit Price'])
data['Total Revenue']=np.where(data['Total Revenue']>=1200,data['Total Revenue'].median(),data['Total Revenue'])
```

```python
future=data.drop('Payment Method',axis=1)
```

```python
target=data['Payment Method']
```

# Splitting the dataset into the training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(future,target,test_size=0.2)
```

```python
X_train
```

```python
y_train
```

# Fitting RANDOM FOREST CLASSIFIER to the training set

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rf=RandomForestClassifier()
```

```python
model=Rf.fit(X_train,y_train)
```

# Predicting the test set results

```python
y_pred=model.predict(X_test)
```

# Machine performance score

```python
model.score(X_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

# Making the confusion matrix

```python
confusion_matrix(y_test,y_pred)
```

```python
X_test
```

# Machine output

```python
def predict(one,two,three,four,five,six,seven,eight,nine):
    for i in model.predict([[one,two,three,four,five,six,seven,eight,nine]]):
        if i==1:
            print('Credit Card')
        elif i==2:
            print('PayPal')
        elif i==3:
            print('Debit Card')
        else:
            pass
```

```python
predict(2,23,2,3232,323232,2,24,4,2024)
```
